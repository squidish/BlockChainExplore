```python
sat = 100000000
```

```python
#can now add up per address
#input a block
def GetOutPutTranaction(block):
    trns_list = block.transactions
    no_trns = block.n_tx
    SentList = []
    RecList = []
    for i in range(0,no_trns):
        nopts = len(block.transactions[i].outputs)
        for j in range(0,nopts):
          if (hasattr(block.transactions[i].outputs[j], 'value') and  hasattr(block.transactions[i].outputs[j], 'address') ):  
            RecList.append([block.transactions[i].outputs[j].value/sat,block.transactions[i].outputs[j].address])
        ninp = len(block.transactions[i].inputs)
        for k in range(0,ninp):
            if (hasattr(block.transactions[i].inputs[k], 'value') and  hasattr(block.transactions[i].inputs[k], 'address') ):
              SentList.append([block.transactions[i].inputs[k].value/sat, block.transactions[i].inputs[k].address])
    return SentList, RecList
```

```python
[SentList, RecList] = GetOutPutTranaction(latest_bock)
```
